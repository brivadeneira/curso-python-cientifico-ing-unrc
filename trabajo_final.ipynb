### Bloque Rojo

#### Breve descripción

El problema en cuestión se reduce a un tiro vertical con posterior caída libre. 

#### Caja negra

*NOTA*: Deberá marcarse con un punto y una leyenda el paso del movimiento de tiro vertical al de caída libre, en los tres gráficos.

#### Sugerencias para la resolución

Para discernir el momento en que se pasa de un tipo de movimiento a otro es útil considerar solamente el tiro vertical, sabemos que la velocidad final en este caso está dada por la fórmula:

$$
v_{final} = v_{inicial} - g.t
$$

Sabiendo que la velocidad final se anula en el punto máximo ($v_{final} = 0$), la expresión se reduce a:

$$
0 = v_{inicial} - g.t_{tv}
$$

de donde se obtiene el tiempo en que se completa el tiro vertical

$$
t_{tv} = \frac{v_{inicial}}{g}
$$

Tanto la posición, velocidad y aceleración del móvil en este punto pueden obtenerse reemplazando en las fórmulas correspondientes:

$$
y = v_{inicial} . t - \frac{1}{2}.g.t^2
$$

$$
v = v_{inicial} - g.t
$$

$$
a = g
$$

Nótese que las expresiones anteriores describen los gráficos a confeccionar:

* $y$ describe una parábola con ramas hacia abajo.
* $v$ una recta con pendiente negativa.
* $a$ una contante.

... formas consistentes con la teoría. 

Solamente queda determinar el tiempo total del movimiento, esto puede hacerse reemplazando $h_1$ en la fórmula de $y$, como sigue:

$$
-h_1 = v_{inicial} . t - \frac{1}{2}.g.t^2
$$

igualando a cero y reordenando

$$
0 = - \frac{1}{2}.g.t^2 + v_{inicial} . t + h_1
$$

De esta expresión pueden aislarse los coeficientes de la ecuación cuadrática necesarios para obtener las raíces mediante la fórmula de Bascara:

$t_{1/2} = \frac {-b \pm \sqrt {b^2 - 4ac}}{2a}$. 

__Es importante la selección de la raiz adecuada de acuerdo a su significado físico.__ 

Una vez obtenido el tiempo total del movimiento queda solamente generar los vectores de $t$, $y$, $v$ y $a$, adecuados para graficar.

La obtención de la velocidad final puede hacerse con un enfoque **cinemático**:

$$
v_{final} = v_{inicial} - g.t_{total}
$$

o mediante consideraciones de **energía**:

$$
K_{inicial}=\frac{1}{2} . m . v_{inicial}^2
$$

$$
U_{inicial} = m.g.h_1
$$

$$
K_{final} = K_{inicial} + U_{inicial}
$$

$$
K_{inicial} + U_{inicial} = \frac{1}{2} . m . v_{final}^2
$$

$$
v_{final}=\sqrt{\frac{2.(K_{inicial}+U_{inicial})}{m}}
$$

De acuerdo al método la velocidad final puede obtenerse con un signo negativo, esto debe ser solucionado a fin de no entorpecer el trabajo del grupo siguiente.

### Bloque Verde

#### Breve descripción

En este bloque la masa asciende por un plano inclinado __con fricción__, pero sin rodar. 

#### Caja negra

**Nombre de la función**: verde

La función debe respetar los siguientes parámetros de entrada y salida:

![](https://i.imgur.com/iM0nnAf.png)

* **Datos de entrada**:
    * Velocidad inicial
    * Masa
    * Altura (indicada en la figura)
    * Ángulo del plano inclinado
    * Coeficiente de fricción del plano

* **Datos de salida**
    * Velocidad final
    * Variable booleana que indique si la masa puede o no ascender el plano
    * Gráficos (en función del tiempo):
        * Posición
        * Velocidad
        * Aceleración

#### Sugerencias para la resolución

Un primer paso para la resolución puede ser calcular la **longitud** $d$ de la hipotenusa del plano inclinado.

$$
\sin{\varphi} = \frac{h_2}{d}
$$

$$
d = \frac{h_2}{\sin{\varphi}}
$$

Un análisis de la **energía** necesaria para ascender el plano requiere el cálculo de la energía cinética inicial y la potencial final del móvil

$$
K_{inicial} = \frac{1}{2}.m.v_{inicial}^2
$$

$$
U_{final} = m.g.h_2
$$

También deben considerarse los **trabajos** realizados por las fuerza de fricción y el peso (en el eje $x$  paralelo al plano)

$$
p_x=m.g.\sin{\varphi}
$$

$$
f_{fricción} = {\mu} .N
$$

$$
N = m.g.\cos{\varphi}
$$

$$
f_{fricción} = {\mu} .m.g.\cos{\varphi}
$$

la fuerza total en el plano es:

$$
f_{total} = p_x + f_{fricción}
$$

el trabajo realizado por estas fuerzas sería:

$$
w_{peso} = p_x . d
$$

$$
w_{fricción} = f_{fricción} . d 
$$

a partir de lo calculado evaluar la posibilidad del ascenso de la masa consiste en **comparar la energía inicial, la final y el trabajo** realizado por las fuerzas en el plano

$$
K_{inicial} - W_{fricción} > U_{final}
$$

El movimiento de la masa en el plano no es más que un **MRUV**, la aceleración puede calcularse por la segunda ley de Newton

$$
f_{total} = m.a
$$

$$
a=\frac{f_{total}}{m}
$$

La duración del movimiento puede deducirse conociendo la distancia final ($d$), la velocidad inicial y la aceleración

$$
d=v_{inicial} . t - \frac{1}{2}.a.t^2
$$

De esta expresión pueden aislarse los coeficientes de la ecuación cuadrática necesarios para obtener las raíces mediante la fórmula de Bascara:

$t_{1/2} = \frac {-b \pm \sqrt {b^2 - 4ac}}{2a}$. 

Debe prestarse atención al significado físico de las soluciones a la ecuación anterior.

La velocidad final puede obtenerse tanto a partir de la fórmula de velocidad 

$$
v=v_{inicial}-a.y
$$


o por consideraciones de energía
$$
K_{final} = K_{inicial} - W_{fricción} - U_{final} = \frac{1}{2}.m.v_{final}^2
$$


$$
v_{final}=\sqrt{\frac{2.K_{final}}{m}}
$$

#### Sugerencias para la resolución

El esquema que sigue aclara la relación entre la altura ganada por el móvil, la longitud de la cadena y el ángulo máximo

![](https://i.imgur.com/BO3vNRx.png)


Por otro lado, un planteo de las **energías** en el punto más bajo y más alto del movimiento de la masa permitiría calcular $h$.

$$
K_{nicial}=\frac{1}{2}.m.v_{inicial}^2
$$

$$
U_{final}=K_{inicial}
$$

$$
U_{final}=m.g.h
$$

$$
h=\frac{U_{final}}{m.g}
$$

$$
h=\frac{\frac{1}{2}.m.v_{inicial}^2}{m.g}

y atendiendo a la figura anterior la **amplitud máxima** puede calcularse por trigonometría

$$
\cos{(\alpha)} = \frac{l-h}{l}

$$
\alpha = \cos^{-1}(\frac{l-h}{l})
$$


El **período** puede calcularse por las dimensiones del péndulo y la gravedad del planeta

$$
\tau = 2.\pi.\sqrt{\frac{l}{g}}
$$

Con esto ya pueden obtenerse las tres expresiones que describan **ángulo, velocidad y aceleración** en función del tiempo:

$$
A (t) = h_{máximo} .\sin{\frac{2.\pi}{\tau}.t}
$$

$$
\omega (t) = h_{máximo} .\cos{\frac{2.\pi}{\tau}.t}
$$

$$
\alpha (t) = h_{máximo} .-\sin{\frac{2.\pi}{\tau}.t}

#### Sugerencias para la resolución

Si se observa el diagrama con detenimiento puede llegarse a una relación entre $r$, $\delta x$, $x_{inicial}$ y $h_3$

$$
2.r=(x-\delta x)+h_3
$$

$$
h_3=2.r-(x- \delta x)
$$

Con esta relación establecida pueden plantearse condiciones de **energía inicial y final**:

$$
K_{inicial} = U_g + U_e
$$

$$
K_{inicial} = m.g.h_3 + \frac{1}{2}.k. \delta x^2
$$

$$
K_{inicial} = m.g.(2.r-x+ \delta x) + \frac{1}{2}.k. \delta x^2
$$

$$
K_{inicial}=m.g.2r-m.g.x+m.g. \delta x + \frac{1}{2}.k. \delta x^2
$$

$$
0 = (\frac{1}{2}.k). \delta x^2 + (m.g). \delta x + (m.g.2r-m.g.x-K_{inicial})
$$

De esta expresión pueden aislarse los coeficientes de la ecuación cuadrática necesarios para obtener las raíces mediante la fórmula de Bascara:

$\delta x = \frac {-b \pm \sqrt {b^2 - 4ac}}{2a}$.

Expresión a partir de la cual puede obtenerse dos posibles valores de $\delta x$, de los cuales debe escogerse el que tenga sentido en el ámbito del problema. 

La altura que alcanza el móvil ya había quedado determinada:

$$
h_3=2.r-(x- \delta x)
$$

Para la confección de los gráficos se debe calcular el tiempo total del movimiento, esto puede lograrse con el planteo de $y_{total}$

$$
y = h_0 + v_{inicial} .t - \frac{1}{2} .g .t^2
$$

$$
2.r-x = r + v_{Final Rampa} . t - \frac{1}{2}.g.t^2
$$

Nótese que se toma como $V_{inicial}$ a la $V_{FinalRampa}$, velocidad que puede calcularse como sigue

$$
K_0 = K_{FinalRampa} + U_{FinalRampa}
$$

$$
\frac{1}{2}.m.v_{inicial} = \frac{1}{2}.m.v_{FinalRampa}^2+m.g.r
$$

$$
v_{FinalRampa} = \sqrt{\frac{2.(\frac{1}{2}.m.v_{inicial}^2-m.g.r)}{m}}
